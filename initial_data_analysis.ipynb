```python
# ALWAYS remove rows where Yield is null (can't train without target)
df = df[df['Yield'].notna()]
print(f"After removing null Yield: {len(df)} rows")
```

```python
df = df.dropna(subset=[
    'Jun-Sep_rain', 'Oct-Dec_rain', 'Jan-Feb_rain', 'Mar-May_rain',
    'Annual_mean_temp', 'Summer_temp', 'Monsoon_temp', 'Winter_temp'
])

df.shape
```

```python
num_cols = [
    'Yield', 'Annual_Rainfall', 'Fertilizer', 'Pesticide',
    'Annual_mean_temp', 'Summer_temp', 'Monsoon_temp', 'Winter_temp',
    'Jun-Sep_rain', 'Oct-Dec_rain', 'Jan-Feb_rain', 'Mar-May_rain'
]
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,8))
sns.boxplot(data=df[num_cols])
plt.xticks(rotation=45)
plt.title("Boxplot to Detect Outliers in Numerical Features")
plt.show()
```

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1
outlier_mask = (df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))
outlier_counts = outlier_mask.sum().sort_values(ascending=False)
outlier_counts
```

```python
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Create mask for outliers
outlier_mask = (df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))

# Drop any row that has an outlier in at least one numeric column
df_no_outliers = df[~outlier_mask.any(axis=1)]
```
